# Data preparation

```python
# Prepared arrays [data, prediction]

FORWARD_STEPS = 90 # the number of values used for 1 prediction
PREDICTION_STEP = 1

def convert_to_Xy_1d(data, feature_idx, n_input=FORWARD_STEPS, n_output=PREDICTION_STEP):
    X, y = [],[]
    in_start = 0
    for i in range(len(data)):
        in_end = in_start + n_input
        out_end = in_end + n_output
        if out_end < len(data):
            x_input = data[in_start:in_end, feature_idx]
            x_input = x_input.reshape((len(x_input), 1)) # column vector
            X.append(x_input)
            y.append(data[in_end:out_end, feature_idx])
        # step forward one unit
        in_start += 1
        i += 1
    return np.array(X), np.array(y)
```

```python
# Reconvert predicted data after applying the model

def comvert_to_array_after_prediction(seq_array, forward_steps = FORWARD_STEPS):
    #seq_result = np.zeros((forward_steps,), dtype=int)
    seq_result=[]
    #seq_result = seq_result.tolist()
    for i in range(0, len(seq_array)):
        seq_result.append(seq_array[i][0])
    return seq_result
```

# LSTM model with seq step 

### LOAD MODEL LSTM_SEQ_MODEL

```python
from keras.models import model_from_json


# load json and create model
json_file = open('model_14_10_100_10_85_78.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
lstm_seq_model_2 = model_from_json(loaded_model_json)
# load weights into new model
lstm_seq_model_2.load_weights("model_14_10_100_10_85_78.h5")
print("Loaded model from disk")
```

# LSTM seg with 2 LSTM layers

```python
trainX_lstm2_seq = trainX_lstm_seq.copy()
testX_lstm2_seq = testX_lstm_seq.copy()

trainy_lstm2_seq = trainy_lstm_seq.copy()
testy_lstm2_seq = testy_lstm_seq.copy()
```

```python
lstm_seq_model_2 = Sequential()
lstm_seq_model_2.add(LSTM(14, activation='relu', return_sequences=True, input_shape=(1, FORWARD_STEPS)))
lstm_seq_model_2.add(LSTM(12, activation='relu'))

lstm_seq_model_2.add(Dense(100))
lstm_seq_model_2.add(Dropout(0.1))
lstm_seq_model_2.add(Dense(10))

lstm_seq_model_2.add(Dense(PREDICTION_STEP))

lstm_seq_model_2.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

#early_stop = EarlyStopping(monitor='loss', patience=3, verbose=1)
history = lstm_seq_model_2.fit(trainX_lstm2_seq, trainy_lstm2_seq, epochs=1000, batch_size=100, verbose=1) #, callbacks=[early_stop], shuffle=False)
```

```python
y_test_pred_lstm_2_seq = lstm_seq_model_2.predict(testX_lstm2_seq)
y_train_pred_lstm_2_seq = lstm_seq_model_2.predict(trainX_lstm2_seq)

y_test_pred_lstm_2_seq[y_test_pred_lstm_2_seq<0]=0
y_train_pred_lstm_2_seq[y_train_pred_lstm_2_seq<0]=0
```

```python
print("The R2 score on the Train set is:\t{:0.3f}".format(r2_score(trainy_lstm2_seq, y_train_pred_lstm_2_seq)))
print("The R2 score on the Test set is:\t{:0.3f}".format(r2_score(testy_lstm2_seq, y_test_pred_lstm_2_seq)))
```
